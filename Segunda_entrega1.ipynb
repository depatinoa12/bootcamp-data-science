```python
Monto=tabla['loan_amount'].unique()
Monto
```

En el siguiente codigo se uso las categorias que son los tipos de credito y en valores  los montos para ver como se comportan los rangos, se observa que hay demasiados valores atipicos. seria interesante ver mas sobre como analizar un boxplot

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax.boxplot([tabla[tabla['credit_type'] == tipo_credito]['loan_amount'] for tipo_credito in tipo_credito], showfliers=True)
ax.set_title('Boxplot de préstamos por tipo de crédito')
ax.set_xlabel('Tipo de crédito')
ax.set_ylabel('Monto del préstamo')
plt.xticks(range(1, len(tipo_credito) + 1), tipo_credito, rotation=45)
plt.show()
```
